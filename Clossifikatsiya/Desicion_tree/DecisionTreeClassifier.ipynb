```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# 1-daraxt: Sun'iy ma'lumotlar uchun
# Ma'lumotlar
X1 = pd.DataFrame({
    'Tajriba': [0.1, 0.4, 0.3, 0.5, 0.6, 0.8, 0.9],
    'Yosh': [0.2, 0.5, 0.6, 0.3, 0.7, 1.0, 0.8]
})
y1 = ['Low', 'Low', 'Medium', 'Low', 'Medium', 'Medium', 'Medium']

# Model
tree1 = DecisionTreeClassifier(max_depth=3, random_state=0)
tree1.fit(X1, y1)

# 2-daraxt: California Uy Narxlari uchun
# Ma'lumotlar
X2 = pd.DataFrame({
    'MedInc': [2.0, 3.0, 5.0, 6.0, 8.0, 2.5, 7.0],
    'HouseAge': [15, 25, 35, 45, 20, 30, 40]
})
y2 = ['Low', 'Low', 'Medium', 'High', 'High', 'Low', 'Medium']

# Model
tree2 = DecisionTreeClassifier(max_depth=4, random_state=0)
tree2.fit(X2, y2)

# Chizish
fig, axs = plt.subplots(2, 1, figsize=(12, 14))

plot_tree(tree1, feature_names=X1.columns, class_names=tree1.classes_, filled=True, ax=axs[0])
axs[0].set_title("Klassifikatsiya Daraxti: Sun'iy Ma'lumotlar")

plot_tree(tree2, feature_names=X2.columns, class_names=tree2.classes_, filled=True, ax=axs[1])
axs[1].set_title("Klassifikatsiya Daraxti: California Uy Narxlari")

plt.tight_layout()
plt.show()
```
