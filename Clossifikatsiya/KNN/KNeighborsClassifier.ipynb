```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Ma'lumotlar: [Yosh, Ta'lim darajasi, Tajriba, Kasb]
data = np.array([
    [25, 4, 2, 0],   # Dasturchi
    [30, 5, 5, 2],   # Menejer
    [22, 3, 1, 0],   # Dasturchi
    [28, 4, 3, 1],   # Dizayner
    [35, 5, 8, 2],   # Menejer
    [24, 3, 2, 0],   # Dasturchi
    [32, 4, 6, 1],   # Dizayner
    [27, 4, 4, 1],   # Dizayner
    [40, 5, 10, 2],  # Menejer
    [26, 3, 3, 0]    # Dasturchi
])

X = data[:, :-1]  # Xususiyatlar (Yosh, Ta'lim, Tajriba)
y = data[:, -1]   # Target (Kasb)

# Modelni yaratamiz (K=3)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)  # Butun ma'lumotlar bilan trenirovka

# Yangi odam haqida bashorat qilamiz
new_person = [[28, 4, 4]]  # Yosh 28, Ta'lim 4, Tajriba 4 yil
prediction = knn.predict(new_person)

# Natijani chiqaramiz
kasblar = {0: "Dasturchi", 1: "Dizayner", 2: "Menejer"}
print(f"Bashorat qilingan kasb: {kasblar[prediction[0]]}")

# Modelni baholaymiz (cross-validation bilan yaxshiroq bo'lardi)
y_pred = knn.predict(X)
accuracy = accuracy_score(y, y_pred)
print(f"Modelning aniqligi: {accuracy * 100:.2f}%")
```
