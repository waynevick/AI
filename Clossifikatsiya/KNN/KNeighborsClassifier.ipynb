```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

data = np.array([
    [25, 4, 2, 0],   # Dasturchi
    [30, 5, 2, 2],   # Menejer
    [29, 4, 1, 0],   # Dasturchi
    [26, 3, 0, 0],   # Dasturchi
    [35, 5, 3, 2],   # Menejer
    [32, 4, 2, 0],   # Dasturchi
    [28, 4, 3, 1],   # Dizayner
    [33, 4, 4, 1],   # Dizayner
    [27, 3, 1, 1],   # Dizayner
    [30, 5, 0, 2],   # Menejer
    [31, 4, 2, 0],   # Dasturchi
])
X = data[:, :-1]
y = data[:, -1]

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)

new_person = [[28, 4, 4]]
prediction = knn.predict(new_person)

kasblar = {0: "Dasturchi", 1: "Dizayner", 2: "Menejer"}
print(f"Bashorat qilingan kasb: {kasblar[prediction[0]]}")

y_pred = knn.predict(X)
accuracy = accuracy_score(y, y_pred)
print(f"Modellning aniqligi: {accuracy * 100:.2f}%")
```
