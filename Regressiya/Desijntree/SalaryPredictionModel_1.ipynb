```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor, plot_tree

# 1-Qaror daraxti
X1 = np.array([[70], [80], [85], [90], [95], [100]])
y1 = np.array([15000, 20000, 25000, 45000, 50000, 60000])
model1 = DecisionTreeRegressor(max_depth=3)
model1.fit(X1, y1)
plt.figure(figsize=(10, 6))
plot_tree(model1, filled=True, feature_names=["House Size"], rounded=True)
plt.title("Decision Tree 1")
plt.show()
```
